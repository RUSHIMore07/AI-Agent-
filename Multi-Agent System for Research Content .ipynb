```python
import warnings
warnings.filterwarnings('ignore')
from crewai import Agent, Task, Crew
import os
from utils import get_openai_api_key
from IPython.display import Markdown
```

```python
openai_api_key = get_openai_api_key()
os.environ["OPENAI_MODEL_NAME"] =  "gpt-4o-mini"

# Research-Focused Planner
planner = Agent(
    role="Research Strategist",
    goal="Develop comprehensive research-backed content plans on {topic} with current data and expert insights",
    backstory="You're a senior research strategist with a background in "
              "academic publishing and data analysis. "
              "Your expertise lies in identifying cutting-edge research "
              "and translating complex information into actionable content plans. "
              "You specialize in finding unique angles and "
              "backing claims with credible sources. "
              "Your research forms the foundation for "
              "creating authoritative content on {topic}.",
    allow_delegation=False,
    verbose=True
)

# Research Content Writer
writer = Agent(
    role="Research Content Writer",
    goal="Transform complex research insights into engaging, "
         "authoritative content about {topic} while maintaining "
         "academic integrity and factual accuracy",
    backstory="You're a seasoned science and research writer "
              "with experience in both academic publishing "
              "and science communication. Your expertise lies "
              "in translating complex research findings into "
              "compelling narratives that engage readers while "
              "preserving scientific accuracy. You have a strong "
              "background in research methodology and data "
              "interpretation, allowing you to effectively "
              "communicate technical concepts to diverse audiences. "
              "You excel at weaving together multiple research "
              "sources to create comprehensive, evidence-based "
              "content that both educates and engages.",
    allow_delegation=False,
    verbose=True
)

# Research Publication Editor
editor = Agent(
    role="Research Publication Editor",
    goal="Ensure content accuracy, research integrity, and scholarly standards while maintaining readability",
    backstory="You're a senior editor with extensive experience "
              "in academic publishing and science communication. "
              "Your expertise includes verifying research claims, "
              "maintaining academic rigor, and ensuring proper "
              "citation of sources while making complex topics "
              "accessible to the target audience. "
              "You excel at balancing scientific accuracy "
              "with engaging narrative flow and have a keen eye "
              "for maintaining consistency in technical terminology "
              "and research presentation.",
    allow_delegation=False,
    verbose=True
)

# Planning Task
plan = Task(
    description=(
        "1. Conduct thorough research on {topic} focusing on:\n"
            "   - Latest research findings and breakthroughs\n"
            "   - Key academic sources and expert opinions\n"
            "   - Current debates and controversies\n"
            "   - Practical applications and implications\n"
        "2. Identify target audience needs and knowledge level.\n"
        "3. Create a detailed content outline with:\n"
            "   - Research context and background\n"
            "   - Methodology overview\n"
            "   - Key findings and analysis\n"
            "   - Practical implications\n"
        "4. Compile essential references and citations.\n"
        "5. Suggest data visualization opportunities."
    ),
    expected_output="A comprehensive research plan including "
                    "detailed outline, key sources, methodology "
                    "approach, and visualization suggestions.",
    agent=planner
)



# Writing Task
write = Task(
    description=(
        "1. Create a comprehensive research-based article using "
            "the detailed outline and sources from the Research Strategist.\n"
        "2. Begin with an engaging introduction that presents "
            "the research context and significance of {topic}.\n"
        "3. Structure the content with these key elements:\n"
            "   - Background and context of the research\n"
            "   - Current state of knowledge in the field\n"
            "   - Key findings and breakthroughs\n"
            "   - Practical implications and applications\n"
            "   - Future directions and open questions\n"
        "4. Include relevant statistics, research findings, and "
            "expert opinions with proper citations.\n"
        "5. Integrate data visualizations, charts, or graphics "
            "where appropriate to illustrate complex concepts.\n"
        "6. Maintain academic rigor while using clear, "
            "accessible language for a broader audience.\n"
        "7. Address potential limitations, alternative viewpoints, "
            "and areas of ongoing research.\n"
        "8. Conclude with actionable insights and broader "
            "implications of the research.\n"
        "9. Incorporate SEO keywords naturally while prioritizing "
            "research accuracy and readability.\n"
        "10. Follow academic style guidelines while ensuring "
            "narrative engagement.\n\n"
        "Format: markdown\n"
        "Style: academic yet accessible\n"
        "Length: comprehensive coverage of topic\n"
        "Audience: educated professionals and interested general readers"
    ),
    expected_output=(
        "A polished, research-based article that combines scholarly "
        "depth with accessible presentation. The article should "
        "include:\n"
        "- Clear section headings and logical flow\n"
        "- Properly cited research findings and expert quotes\n"
        "- Data visualizations or explanatory elements\n"
        "- 2-3 well-developed paragraphs per section\n"
        "- Academic citations in appropriate format\n"
        "- A balance of technical detail and reader engagement\n"
        "The final piece should be ready for expert review while "
        "remaining accessible to an educated general audience."
    ),
    agent=writer
)


# Editing Task
edit = Task(
    description=(
        "1. Review the content for academic and research accuracy, "
            "ensuring all claims are properly supported.\n"
        "2. Verify citations and references, confirming they are "
            "current and authoritative sources.\n"
        "3. Check for proper use of technical terminology and "
            "consistency in research presentation.\n"
        "4. Ensure complex concepts are explained clearly without "
            "oversimplification.\n"
        "5. Review data presentation and statistical accuracy.\n"
        "6. Maintain balance between academic rigor and "
            "reader accessibility.\n"
        "7. Assess the logical flow of research arguments "
            "and supporting evidence.\n"
        "8. Ensure compliance with academic style guidelines "
            "while preserving engagement."
    ),
    expected_output="A thoroughly vetted, academically sound article "
                    "that maintains research integrity while being "
                    "accessible and engaging. The final piece should "
                    "include proper citations, clear methodology "
                    "explanations, and well-supported conclusions "
                    "presented in a reader-friendly format.",
    agent=editor
)

# Creating the Research-Focused Crew
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2
)
# Example usage
result = crew.kickoff(inputs={"topic": "The Impact of Artificial Intelligence on Healthcare"})
```
